    <h2 class="title_level_2" style="margin-left:17%;font-family:Georgia, serif">4. ESTRUCTURAR CELDA TIPO MARKDOWN</h2>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Una celda es un espacio de trabajo el cual puede ser de diferentes tipos, para el ejemplo se ha seleccionado una celda tipo
        MARKDOWN
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Markdown es un lenguaje de marcado con sintaxis en texto plano para generar texto con formato.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Al utilizar MARKDOWN se pueden producir archivos que pueden ser legibles como texto plano y que a la vez están listos para
        ser formados en otras plataformas.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Por lo tanto al utilizar las celdas de tipo MARKDOWN se puede hacer uso de su lenguaje junto con HTML para la presentación
        de textos, gráficas, ecuaciones y un gran número de elementos más para dar vida al contenido que se quiere mostrar.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        A continuación se describe el diseño que debe llevar la celda de tipo MARKDOWN para los notebooks creados por los estudiantes
        y docentes del ITM.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        El contenido de cada celda deberá llevar una etiqueta div de HTML que sea el componente padre de todo el contenido, alli
        se deberán incluir las demas etiquetas ya sea HTML o MARKDOWN que sean necesarias para enriquecer la experiencia
        visual del lector.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Aunque el div padre no es una camisa de fuerza ya que se podrían utilizar una gran cantidad de etiquetas HTML o MARKDOWN
        para dar inicio al documento, es recomendable llevar un estándar que proporcione un orden para facilidad en el mantenimiento
        de los notebooks.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Como segunda recomendación se sugiere que el contenido de la celda no sea extensa en cuanto a etiquetas, eso no quiere decir
        que no se pueda poner un texto largo, se puede hacer pero evitando usar muchas etiquetas que dificulten el mantenimiento
        del notebook.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Como tercera recomendación se sugiere el uso de indentación para facilidad en la lectura del código, tal como se ve en la
        imagen.
    </p>
</div>

    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Los párrafos se deben crear con la etiqueta
        <b>p</b> de html y se sugiere que lleven las reglas de un párrafo como por ejemplo la longitud del texto. Todos los
        párrafos deben llevar el mismo estilo para conservar las márgenes y su tipo de letra, para aquellas palabras que
        necesiten énfasis se puede utilizar la etiqueta &lt;b&gt; (bold) o
        <b>strong</b> para resaltar en negrilla.
    </p>
    <ul class="unordered-list" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">
        <li>etiqueta: p</li>
        <li>class: main-p</li>
        <li>Estilo css:
            <span style="color:#4169E1">.main-p{ margin-left:auto; margin-right:auto; margin-bottom:24px; font-family:Georgia, serif; width:66%; line-height:
                170%; }
            </span>
        </li>
        <li>Componente:
            <br>
            <span style="color:#4169E1">
                &lt;p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia, serif;width:66%;line-height:
                170%;"&gt; &lt;/p&gt;
            </span>
        </li>
        <li>Ejemplo visual:
            <br>
            <div class="img-div" style="margin-left:0%;width:95%;">
                <img src="res/handbook-img-19.png">
            </div>
        </li>
    </ul>

    <h3 class="title_level_3" style="margin-left:17%;font-family:Georgia, serif">5.4 Imágenes.</h3>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Las imagenes deben de ir dentro de una etiqueta &lt;div&gt; que las contiene y una etiqueta &lt;img&gt; que es la encargada
        de mostrar la imagen, el tamaño de la imagen lo define el creador del notebook según su importancia, pero esta imagen
        nunca sobrepasa el tamaño del &lt;div&gt; contenedor.
    </p>
    <ul class="unordered-list" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">
        <li>etiqueta: img</li>
        <li>class: img-div</li>
        <li>Estilo css:
            <span style="color:#4169E1">.img-div{ margin-left:15%; width:65%; }

            </span>
        </li>
        <li>Componente:
            <br>
            <span style="color:#4169E1">
                &lt;div class="img-div" style="margin-left:15%;width:65%;"&gt; &lt;img src=""&gt;&lt;div&gt;
            </span>
        </li>
         <li>Componente con link:
            <br>
            <span style="color:#4169E1">
                &lt;div class="img-div" style="margin-left:15%;width:65%;"&gt;&lt;a href=""&gt; &lt;img src=""&gt;&lt;/a&gt;&lt;div&gt; 
            </span>
        </li>
