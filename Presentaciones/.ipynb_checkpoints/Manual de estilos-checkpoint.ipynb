    <h2 class="title_level_2" style="margin-left:17%;font-family:Georgia, serif">4. ESTRUCTURAR CELDA TIPO MARKDOWN</h2>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Una celda es un espacio de trabajo el cual puede ser de diferentes tipos, para el ejemplo se ha seleccionado una celda tipo
        MARKDOWN
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Markdown es un lenguaje de marcado con sintaxis en texto plano para generar texto con formato.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Al utilizar MARKDOWN se pueden producir archivos que pueden ser legibles como texto plano y que a la vez están listos para
        ser formados en otras plataformas.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Por lo tanto al utilizar las celdas de tipo MARKDOWN se puede hacer uso de su lenguaje junto con HTML para la presentación
        de textos, gráficas, ecuaciones y un gran número de elementos más para dar vida al contenido que se quiere mostrar.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        A continuación se describe el diseño que debe llevar la celda de tipo MARKDOWN para los notebooks creados por los estudiantes
        y docentes del ITM.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        El contenido de cada celda deberá llevar una etiqueta div de HTML que sea el componente padre de todo el contenido, alli
        se deberán incluir las demas etiquetas ya sea HTML o MARKDOWN que sean necesarias para enriquecer la experiencia
        visual del lector.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Aunque el div padre no es una camisa de fuerza ya que se podrían utilizar una gran cantidad de etiquetas HTML o MARKDOWN
        para dar inicio al documento, es recomendable llevar un estándar que proporcione un orden para facilidad en el mantenimiento
        de los notebooks.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Como segunda recomendación se sugiere que el contenido de la celda no sea extensa en cuanto a etiquetas, eso no quiere decir
        que no se pueda poner un texto largo, se puede hacer pero evitando usar muchas etiquetas que dificulten el mantenimiento
        del notebook.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Como tercera recomendación se sugiere el uso de indentación para facilidad en la lectura del código, tal como se ve en la
        imagen.
    </p>
</div>

    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        En caso de que se requiera utilizar títulos con fuentes más pequeñas, se puede utilizar las etiquetas &lt;h4&gt;, &lt;h5&gt;
        utilizando el mismo estilo de la etiqueta &lt;h3&gt;

    <h3 class="title_level_3" style="margin-left:17%;font-family:Georgia, serif">5.3 Párrafos.</h3>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Los párrafos se deben crear con la etiqueta
        <b>p</b> de html y se sugiere que lleven las reglas de un párrafo como por ejemplo la longitud del texto. Todos los
        párrafos deben llevar el mismo estilo para conservar las márgenes y su tipo de letra, para aquellas palabras que
        necesiten énfasis se puede utilizar la etiqueta &lt;b&gt; (bold) o
        <b>strong</b> para resaltar en negrilla.
    </p>
    <ul class="unordered-list" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">
        <li>etiqueta: p</li>
        <li>class: main-p</li>
        <li>Estilo css:
            <span style="color:#4169E1">.main-p{ margin-left:auto; margin-right:auto; margin-bottom:24px; font-family:Georgia, serif; width:66%; line-height:
                170%; }
            </span>
        </li>
        <li>Componente:
            <br>
            <span style="color:#4169E1">
                &lt;p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia, serif;width:66%;line-height:
                170%;"&gt; &lt;/p&gt;
            </span>
        </li>
        <li>Ejemplo visual:
            <br>

    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Las imagenes deben de ir dentro de una etiqueta &lt;div&gt; que las contiene y una etiqueta &lt;img&gt; que es la encargada
        de mostrar la imagen, el tamaño de la imagen lo define el creador del notebook según su importancia, pero esta imagen
        nunca sobrepasa el tamaño del &lt;div&gt; contenedor.

         <li>Componente con link:

    <h3 class="title_level_3" style="margin-left:17%;font-family:Georgia, serif">5.5 Fórmulas de una sola columna.</h3>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Las formulas son parte esencial de los notebooks, por lo tanto llevan un estilo propio ya que pueden ser incluidas dentro
        de un párrafo y deben ser diferenciables del texto simple, estas fórmulas deben ser centradas con la etiqueta &lt;center&gt;
        de html e incluidas en una etiqueta &lt;span&gt;.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Para las fórmulas se ha utilizado Latex, que es un lenguaje de programación de computadoras usado para tipografia de documentos
        tecnicos y cientificos.
    </p>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Para las personas que no manejan Latex, se puede hacer uso de una pagina web para la generación de código Latex ingresando

    <h3 class="title_level_3" style="margin-left:17%;font-family:Georgia, serif">5.6 Algoritmos.</h3>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Otra parte esencial de los notebooks son los algoritmos, para ello se ha creado un estilo apropiado para diferenciarlo del texto simple o de las fórmulas, como normalmente los algoritmos van creciendo verticalmente, se ha optado por utilizar la etiqueta &lt;ul&gt; de HTML que cumple la función de lista, y para generar la indentación, a cada uno de los &lt;li&gt; (filas) se le ha agregado una margen izquierda (margin-left) que va aumentando o disminuyendo de a 30px según la necesidad.
    </p>        
    <ul class="unordered-list" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">
        <li>etiqueta: ul</li>
        <li>class: algorithm</li>
        <li>Estilo css:
            <span style="color:#4169E1">.algorithm{
   display:block;
   margin-left:20%;
}
<br>
.algorithm ul{
   list-style:none;
   font-family:cursive;
   color:SteelBlue;   
}
<br>
.algorithm ul li{
   margin-top:6px
}


            </span>
        </li>
        <li>Componente:
            <br>
            <span style="color:#4169E1">
                &lt;div class="algorithm" style="display:block;margin-left:20%"&gt;
  &lt;ul style="list-style:none;font-family:cursive;color: SteelBlue"&gt;
       &lt;li style="margin-left:0px;margin-top:6px">For i = 0 to m {&lt;/li&gt;
       &lt;li style="margin-left:30px;margin-top:6px">if ... {&lt;/li&gt;
       &lt;li style="margin-left:60px;margin-top:6px">...&lt;/li&gt;
       &lt;li style="margin-left:30px;margin-top:6px">}&lt;/li&gt;
       &lt;li style="margin-left:0px;margin-top:6px">}&lt;/li&gt;
  &lt;/ul&gt;
&lt;/div&gt;
            </span>

    <h3 class="title_level_3" style="margin-left:17%;font-family:Georgia, serif">5.7 Componentes en doble columna.</h3>
    <p class="main-p" style="margin-left:auto;margin-right:auto;margin-bottom:24px;font-family:Georgia,        serif;width:66%;line-height: 170%;">
        Hay casos en que se quiere ver dos fórmulas, dos imágenes, dos textos, etc. en una misma fila pero separados por columnas,
        para estos casos se tiene un componente que hace que esto sea posible sin necesidad de modificar los componentes
        que ya se ha visto en este manual, simplemente se coge el componente que se quiere y se ubica dentro del actual.
    </p>
    <ul class="unordered-list" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">

                EL PRIMER COMPONENTE QUE QUEREMOS &lt;/div&gt; &lt;div class="column" style="float: left;width: 50%;padding:
                10px;height: auto;"&gt; AQUI AGREGAMOS EL SEGUNDO COMPONENTE QUE QUEREMOS &lt;/div&gt; &lt;/div&gt; &lt;/div&gt;

    <h2 class="title_level_2" style="margin-left:17%;font-family:Georgia, serif">6. DEFINICIONES</h2>
    <ol class="ordered-list" start="1" style="margin-left:auto;margin-right:auto;margin-bottom:0px;width:66%;line-height: 170%;color:black;font-family:Georgia, serif">
        <li>
            <strong>CELDA</strong>
            <br> En informática, las celdas son campos de variable tamaño y disposición que permiten introducir datos, en general
            numéricos, para ponerlos en relación y ejecutar operaciones matemáticas [3].
        </li>
        <li>
            <hr>
            <strong>MARKDOWN</strong>
            <br> es una forma sencilla de agregar formato a textos web. Con formato nos referimos a itálicas, negritas, listas,
            y más, en texto plano –por ejemplo, el texto que nos encontramos en el visor de HTML de los editores de blogs,
            o en el Bloc de Notas de Windows, para hacer el concepto más asequible-. Markdown fue creado originalmente por
            John Gruber, con ayuda de Aaron Swartz, con el propósito de crear un texto plano fácil de escribir y fácil de
            leer, y que pudiera convertirse de forma sencilla y válida a XHTML [4].
        </li>
        <li>
            <hr>
            <strong>Raw NBConvert</strong>
            <br> Es un tipo de celda en jupyter notebook que puede ser creada en diferentes tipos de formatos como lo son HTML
            y Latex [5].
        </li>
        <li>
            <hr>
            <strong>Latex</strong>
            <br> es un sistema de composición tipográfica de alta calidad; incluye características diseñadas para la producción de documentación técnica y científica. LaTeX es el estándar de facto para la comunicación y publicación de documentos científicos [6].
        </li>
    </ol>
